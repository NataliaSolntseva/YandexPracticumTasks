Для унификации названий можем сократить название до имени собственного. Для этого создадим функцию `cut_str_before_first_upper`, которая обрезает значения `locality_name` до слова, начинающего с большой буквы и применим ее к столбцу датафрейма при помощи функции `apply`

```python
# создадим функцию cut_str_before_first_upper, обрезающую строку до первого символа в верхнем регистре 

def cut_str_before_first_upper(str_for_cut):
    try: 
        for i in range(len(str_for_cut)):
            if str_for_cut[i].isupper() == True:
                return str_for_cut[i:]       
    except:
        return str_for_cut
```

```python
#  применим функцию cut_str_before_first_upper к столбцу датафрейма
data_clear['locality_name'] = data_clear['locality_name'].apply(cut_str_before_first_upper)
```

###### Функция для подсчета значений усов диагаммы размаха 

**Вспомним правило формирования размаха "усов" диаграммы размаха:**

*«Ящик» ограничен первым и третьим квартилями. Внутри ящика обозначают медиану.
«Усы» простираются влево и вправо от границ ящика на расстояние, равное 1,5 межквартильного размаха (сокращённо IQR, от англ. interquartile range, «межквартильный размах»). 
В размах «усов» попадают нормальные значения, а за пределами находятся выбросы, изображённые точками. Если правый «ус» длиннее максимума, то он заканчивается максимумом. То же — для минимума и левого «уса».*

```python
# напишем функцию, которая:
# - считает значения для усов размаха для столбца pandas
# - выводит на экран их значения 
# - оценивает объем записей за пределами усов 

# data_w - таблица pandas
# col_name - наименование столбца
# x_min - минимальное значение
# q_1 - первый квартиль (25%)
# q_2 - третий квартиль (75%)
# x_max - максимальное значение 

def whisker(data_w, col_name, x_min, q_1, q_2, x_max,):
    try:
        # подсчет межвартильного размаха IQR 
        iqr = q_2 - q_1
        
        # подсчет значения правого уса
        right_w = q_2 + iqr * 1.5
        if right_w > x_max:
            right_w = x_max
        
        # подсчет значения левого уса 
        left_w = q_1 - iqr * 1.5 
        if left_w < x_min:
            left_w = x_min 
        
        # вывод результата на экран
        print('Исходные данные:')
        print('    Межквартильный размах = ', iqr)
        print('    Минимальное значение = ', x_min)
        print('    Первый квартиль (25%) = ', q_1)
        print('    Третий квартиль (75%) = ', q_2)
        print('    Максимальное значение = ', x_max)
        print()
        print('Результаты расчетов:')
        print('    Правый ус = ', right_w)
        print('    Левый ус = ', left_w)
        print('    Количество записей', col_name,' за пределами усов: ', 
              data_w.query(f'{col_name} > @right_w or {col_name} < @left_w').shape[0])
        print('    Процент записей ', col_name,'за пределами уса в % от общего числа: ', 
              data_w.query(f'{col_name} > @right_w or {col_name} < @left_w').shape[0] / data_w.shape[0] * 100)
         
    except:
        print('Что-то пошло не так при рассчетах')
```

Проверим корректность работы функции `whisker` для рассчета усов размаха.

Для этого сначала определим вручную границы усов диаграммы размаха и подчитаем количество записей, выпадающих за пределы усов. 
Затем вызовем функцию и сравним полученные результаты. 

В параметрах статистики для days_exposition мы получили следующие значения:

- min          1.000000
- 25%         45.000000
- 50%         95.000000
- 75%        232.000000
- max       1580.000000

Из них следует, что: 
- межквартильный размах = 232 - 45 = 187 
- правый ус = 232 + 187*1.5 = 512,5 
- левый ус = 45 - 187*1.5 = -235,5 (но т.к. есть минимум то левый ус заканчивается им - это значение 1 )

```python
# посчитаем сколько записей выпадают за пределы усов 
print('Количество записей days_exposition за пределами усов: ', 
      data_clear.query('days_exposition < 1 or days_exposition > 512.5').shape[0])
print('Процент записей от общего числа: ', 
      data_clear.query('days_exposition < 1 or days_exposition > 512.5').shape[0] / data_clear.shape[0] * 100)
```

```python
# проверим работу функции
whisker(data_w=data_clear, 
        col_name='days_exposition', 
        x_min=data_clear['days_exposition'].describe().loc['min'], 
        q_1=data_clear['days_exposition'].describe().loc['25%'], 
        q_2=data_clear['days_exposition'].describe().loc['75%'], 
        x_max=data_clear['days_exposition'].describe().loc['max'])
```

**Функция работает корректно - результаты ее работы совпали с ручным вычислением значений. Можно ее использовать для рассчетов на других параметрах**

Достоверно сказать, в каком именно значении допущена ошибка (в площади кухни, в жилой площади или в общей плоащди) достоверно невозможно. Поэтому не будем исправять эти значения и оставим их как есть.  

Заполним пропуски в `kitchen_area`, использую формулу с поправочным коэффициентом:

```python
# перед заполнением пропусков проверим количество аномалий в соотношении значений, 
# чтобы оценить, сколько пропущенных значений может быть заполнено  

print('Количество записей, в которых kitchen_area==NaN & living_area != NaN: ', 
      data_clear.query('kitchen_area.isna() and not living_area.isna()').shape[0])
print('Количество записей, в которых kitchen_area==NaN & living_area != NaN & (0.9 * total_area - living_area) >= 0: ', 
      data_clear.query('kitchen_area.isna() and not living_area.isna() and (0.9 * total_area - living_area) >= 0')
      .shape[0])
print('Количество записей, в которых kitchen_area==NaN & living_area != NaN & (total_area - living_area) >= 0: ', 
      data_clear.query('kitchen_area.isna() and not living_area.isna() and (total_area - living_area) >= 0')
      .shape[0])
```

```python
'''
напишем функцию fill_kitchen, которая принимает на вход строку датафрема и заполнет значения столбца kitchen_area 
следюущим образом:
 - если kitchen_area пропущено, а living_area нет, то заполнять его как (0.9 * total_area - living_area)
 - при этом, чтобы не получить отрицательных значений в площади кухни, то дополнительно проверять что 
 (0.9 * total_area - living_area) >= 0
 - если значение (0.9 * total_area - living_area) < 0, то применим формулу без коэффицента 0.9, при этом аналогично проверив, 
 что заполнение total_area - living_area не даст отрицательного значения  
'''

def fill_kitchen(row):
    if np.isnan(row['kitchen_area']) and not np.isnan(row['living_area']) and 0.9*row['total_area'] - row['living_area'] >= 0:
        return 0.9*row['total_area'] - row['living_area']
    elif np.isnan(row['kitchen_area']) and not np.isnan(row['living_area']) and row['total_area'] - row['living_area'] >= 0:
        return row['total_area'] - row['living_area']       
    else: 
        return row['kitchen_area']
```

```python
data_clear['kitchen_area'] = data_clear.apply(fill_kitchen, axis=1)

# уникальные значения после заполнения по формуле
print(data_clear['kitchen_area'].sort_values().unique())
```

Заполним пропуски в `living_area`, использую формулу с поправочным коэффициентом:

```python
# перед заполнением пропусков проверим количество аномалий в соотношении значений, 
# чтобы оценить, сколько пропущенных значений может быть заполнено  

print('Количество записей, в которых kitchen_area != NaN & living_area == NaN: ', 
      data_clear.query('not kitchen_area.isna() and living_area.isna()').shape[0])

print('Количество записей, в которых kitchen_area != NaN & living_area == NaN & (0.9 * total_area - kitchen_area) >= 0: ', 
      data_clear.query('not kitchen_area.isna() and living_area.isna() and (0.9 * total_area - kitchen_area) >= 0')
      .shape[0])

print('Количество записей, в которых kitchen_area != NaN & living_area == NaN & (total_area - kitchen_area) >= 0: ', 
      data_clear.query('not kitchen_area.isna() and living_area.isna() and (total_area - kitchen_area) >= 0')
      .shape[0])
```

Результаты проверки показали, что мы можем по формуле заменить все пропуски `living_area`, для которых задано значение `kitchen_area`. И это число пропусков - 1903 - соответствует общему числу пропусков в `living_area`

```python
'''
напишем функцию fill_living, которая принимает на вход строку датафрема и заполнет значения столбца living_area 
следюущим образом:
 - если living_area пропущено, а kitchen_area нет, то заполнять его как (0.9 * total_area - kitchen_area)
 - при этом, чтобы не получить отрицательных значений в жилой площади, то дополнительно проверять что 
 (0.9 * total_area - kitchen_area) >= 0
 - если значение (0.9 * total_area - kitchen_area) < 0, то применим формулу без коэффицента 0.9, при этом аналогично проверив, 
 что заполнение total_area - kitchen_area не даст отрицательного значения  
'''

def fill_living(row):
    if not np.isnan(row['kitchen_area']) and np.isnan(row['living_area']) and 0.9*row['total_area'] - row['kitchen_area'] >= 0:
        return 0.9*row['total_area'] - row['kitchen_area']
    elif not np.isnan(row['kitchen_area']) and np.isnan(row['living_area']) and row['total_area'] - row['kitchen_area'] >= 0:
        return row['total_area'] - row['kitchen_area']       
    else: 
        return row['living_area']
```

```python
data_clear['living_area'] = data_clear.apply(fill_living, axis=1)

# уникальные значения living_area после заполнения по формуле
print(data_clear['living_area'].sort_values().unique())
```

```python
# посмотрим коэффициенты корреляции столбцов датасета для Санкт-Петербурга
display(data.loc[data['locality_name'] == 'Санкт-Петербург'].corr())
```

По коэффициенту корреляции видно, что нет прямой зависимости `floors_total` от каких-либо величин датасета. Единственными условно подходящими значениями могут быть: 
- `cityCenters_nearest` с положительным коэффициетом корреляции 0.359832 - можно предположить, что при удалении от центра города этажность возрастает, учитывая что все многооэтажные новостройки располагаются ближе к окране города, а центр представлен исторческими жилыми объектами с малой этажностью 
- `floor` с положительным коэффициентом корреляции 0.648303 - здесь мы видим условно большой коэффициент, и он достаточно очевиден, ведь чем больше этажей в доме, тем больше квартир расположнных на более высоких этажах 

Но более надежным выглядит следующий подход в заполнении пропусков `floors_total`:
- заполнять пропуски отдельно с первичной группировкой для каждого города 
- заполнять медианой с вторичной группировкой по `cityCenters_nearest`, при этом деляя проверку, что медианное общее количество этажей не меньше этажа объекта: 
    - если превышает, то заполнить медианой с группировкой по городу, при этом сделать аналогичную проверку, что медианная общее количество этажей не меньше этажа объекта: 
        - если меньше, то заполняем пропуск значением этажа объекта  

```python
# для каждого уникального названия населенного пункта, где есть пропуски в floors_total
for ln in data_clear.loc[data_clear['floors_total'].isna(), 'locality_name'].unique():
    
    # для каждой строки датасета где есть пустое значение floors_total и город равен ln
    for ccn in data_clear.loc[(data_clear['floors_total'].isna()), 'cityCenters_nearest'].unique(): 
        
        # найдем медиану floors_total с группировкой по городу и расстоянию до центра 
        ccn_median = data_clear.loc[(data_clear['locality_name'] == ln) &
                                    (data_clear['cityCenters_nearest'] == ccn), 'floors_total'].median()
        # заполнение пропусков медианой ccn_median для тех значений где медиана не меньше этажа объекта 
        data_clear.loc[(data_clear['locality_name'] == ln) &
                       (data_clear['cityCenters_nearest'] == ccn) &
                       (data_clear['floor'] <= ccn_median) &
                       (data_clear['floors_total'].isna()), 'floors_total'] = ccn_median
        
    # найдем медиану floors_total только с группировкой по городу 
    ln_median = data_clear.loc[(data_clear['locality_name'] == ln), 'floors_total'].median()
    # заполнение пропусков медианой ln_median для тех значений где медиана не меньше этажа объекта 
    data_clear.loc[(data_clear['locality_name'] == ln) &
                   (data_clear['floor'] <= ln_median) &
                   (data_clear['floors_total'].isna()), 'floors_total'] = ln_median

print('Количество пропусков floors_total после заполнения медианами:', data_clear['floors_total'].isna().sum())

data_clear.loc[data_clear['floors_total'].isna(), 'floors_total'] = data_clear['floor']

print('Количество пропусков floors_total после заполнения занчением этажа объекта:', data_clear['floors_total'].isna().sum())
```

Следуя такой логике применим следующий подход в заполнении пропусков `ponds_nearest`:
- заполнять пропуски отдельно с первичной группировкой для каждого города 
- заполнять медианой с вторичной группировкой по `ponds_around3000` 
- для оставшихся пустых значений заполним медианой с группировкой по городу 
 

```python
# для каждого уникального названия населенного пункта, где есть пропуски в ponds_nearest
for ln in data_clear.loc[data_clear['ponds_nearest'].isna(), 'locality_name'].unique():
    
    # для каждого уникального кол-ва парков где есть пустое значение ponds_nearest и город равен ln
    for pa in data_clear.loc[(data_clear['ponds_nearest'].isna()), 'ponds_around3000'].unique(): 
        
        # найдем медиану ponds_nearest с группировкой по городу и кол-ву парков 
        pa_median = data_clear.loc[(data_clear['locality_name'] == ln) &
                                    (data_clear['ponds_around3000'] == pa), 'ponds_nearest'].median()
        # заполнение пропусков медианой ccn_median для тех значений где медиана не меньше этажа объекта 
        data_clear.loc[(data_clear['locality_name'] == ln) &
                       (data_clear['ponds_around3000'] == pa) &
                       (data_clear['ponds_nearest'].isna()), 'ponds_nearest'] = pa_median
        
    # найдем медиану ponds_nearest только с группировкой по городу 
    ln_median = data_clear.loc[(data_clear['locality_name'] == ln), 'ponds_nearest'].median()
    # заполнение пропусков медианой ln_median для тех значений где медиана не меньше этажа объекта 
    data_clear.loc[(data_clear['locality_name'] == ln) &
                   (data_clear['ponds_nearest'].isna()), 'ponds_nearest'] = ln_median

print('Количество пропусков ponds_nearest после заполнения медианами:', data_clear['ponds_nearest'].isna().sum())
```

### Столбец `floor_type` (тип этажа квартиры (значения — «первый», «последний», «другой»))

- «первый», 
- «последний», 
- «другой»

```python
# напишем функцию floor_type_func, которая возвращает тип этажа квартиры - «первый», «последний», «другой»
# входным параметром функции является строка датафрейма

def floor_type_func(row):
    try:
        # если этаж объекта равен 1 
        if row['floor'] == 1:
            return 'первый'
        # если этаж объекта совпадает с общей этажностью дома
        elif row['floor'] == row['floors_total']:
            return 'последний'
        # для всех остальных случаев 
        else:
            return 'другой'
    except:
        pass
```

```python
# создадим новый столбец floor_type, который будет содержать месяц публикации объявления 
df_final['floor_type'] = df_final.apply(floor_type_func, axis=1)

# выведем первые 5 строк датафрейма, чтобы пороверить корректность заполнения столбца floor_type
display(df_final[['floor', 'floors_total', 'floor_type']].head())
```

```python
# напишем функцию cnn_km_func, которая возвращает расстояние до центра города в километрах
# входным параметром функции является значение столбца датафрейма

def cnn_km_func(cnn):
    try:
        if not np.isnan(cnn):
            # чтобы перевести метры в километры, поделим число на 1000 округлим по математическим правилам 
            return round(cnn / 1000) 
    except:
        pass
```

Посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделите населённые пункты с самой высокой и низкой стоимостью квадратного метра. Эти данные можно найти по имени в столбце `locality_name`.

```python
# построение столбчатой диаграммы с использованрием pivot_table по топ-10 городов 
# без дополнительной переменной 
(
    df_final
    .pivot_table(index='locality_name', values='one_meter_price', aggfunc=['count', 'median'])
    .sort_values(by=('count','one_meter_price'), ascending=False)
    .head(10)
    .sort_values(by = ('median','one_meter_price'))
    .plot(title='Средняя цена одного квадратного метра в топ-10 городов  \n\n', 
          grid=True, kind='barh', y=('median','one_meter_price'), figsize=(12, 6))
);
```

```python
# первый способ построения столбчатого графика через  groupby
(
    df_final.query('locality_name in @ln_list')
    .groupby('locality_name')['one_meter_price']
    .median()
    .sort_values()
    .plot(title='Средняя цена одного квадратного метра в топ-10 городов  \n\n', grid=True, 
          kind='barh', figsize=(12, 6))
);
```

```python
# второй способ построения столбчатой диаграммы с использованрием pivot_table
(
    df_final.query('locality_name in @ln_list')
    .pivot_table(index='locality_name', values='one_meter_price', aggfunc='median')
    .sort_values(by='one_meter_price')
    .plot(title='Средняя цена одного квадратного метра в топ-10 городов  \n\n', 
          grid=True, kind='barh', figsize=(12, 6))
);
```

```python
# Выведем топ-10 городов с указанием средней стоимости квадратного метра входят (в порядке убывания стоимости):
(
    df_final.query('locality_name in @ln_list')
    .pivot_table(index='locality_name', values='one_meter_price', aggfunc='median')
    .sort_values(by='one_meter_price', ascending=False).round(decimals=2)
)
```

Среди топ-10 городов самая большая цена за квадратный метр в Санкт-Петербурге 103 794, самая дешевая - в Выборге 58 158. Среди всех остальных городов, расположеных со 2-ого по 9-ое место, цены за квадртаный отличаются друг от друга в порядке убывания примерно на одинковую величину около 5 000.
