Датасет содержит следующие данные клиента за месяц, предыдущий до проверки факта оттока:

1) данные о клиенте: 
- `gender` — пол;
- `Near_Location` — проживание или работа в районе, где находится фитнес-центр;
- `Partner` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);

- `Phone` — наличие контактного телефона;
- `Age` — возраст;
- `Lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).


2) данные на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
- `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
- `Group_visits` — факт посещения групповых занятий;
- `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
- `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
- `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

3) данные об оттоке клиента
- `Churn` — факт оттока в текущем месяце

Датасет включает 4000 записей без пропусков. Типы данных для всех столбцов соответсвуют хранимым в них значениях, за исключением `Month_to_end_contract` - он хранил данные в десятичном формате и поскольку он содержал целоцисленные значения месяца, то его тип был изменен на `int`

Построим столбчатые диаграммы со средними значениями количественных признаков в разбивке по признаку оттока

```python
# check
#df_gym.groupby('Churn').agg({'Age': 'mean'}).reset_index().round(decimals=2)
```

```python
# подсчитаем количество строк для построения общего графика
# на одной строке будет размещего по 4 графика 
count_plt_row = math.ceil(len(list_no_bin_f)/4)

plt.figure(figsize=(15, 5 * count_plt_row)).suptitle('Средние значения категориальных принзаков в разбивке по оттоку', 
                                                     fontsize=16)

# перебираем элементы списка и для каждого считаем отдельный барплот 
for counter, i in enumerate(list_no_bin_f):
        
    ax = plt.subplot(count_plt_row, 4, counter+1)
    
    ax1 = sns.barplot(data=df_gym.groupby('Churn').agg({i: 'mean'}).reset_index().round(decimals=2),
          y=i, x='Churn')
    
    # устанавливаем подписи на бары - ФУНКЦИЯ НЕ РАБОТАЕТ В ОКРУЖЕНИИ ЯНДЕКСА, 
    # здесь и далее пришлось закомментить
    for container in ax1.containers:
        ax1.bar_label(container)
    
    plt.title('Среднее значения признака \n {}'.format(i), fontsize=14, pad=5)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.xlabel('Факт оттока')
    plt.ylabel('Среднее значение признака')

plt.tight_layout(w_pad=1)
plt.subplots_adjust(top=0.9) 
plt.show()
```

Дополнительно построим столбчатые диаграммы с медианными значениями количественных признаков в разбивке по признаку оттока. Т.к. медиана не чувствительна к выбросам, то этот параметр более показателен

```python
# подсчитаем количество строк для построения общего графика
# на одной строке будет размещего по 4 графика 
count_plt_row = math.ceil(len(list_no_bin_f)/4)

plt.figure(figsize=(15, 5 * count_plt_row)).suptitle('Медианные значения категориальных принзаков в разбивке по оттоку', 
                                                     fontsize=16)

# перебираем элементы списка  и для каждого считаем отдельный барплот 
for counter, i in enumerate(list_no_bin_f):
        
    ax = plt.subplot(count_plt_row, 4, counter+1)
    
    ax1 = sns.barplot(data=df_gym.groupby('Churn').agg({i: 'median'}).reset_index().round(decimals=2),
          y=i, x='Churn')
    
    # устанавливаем подписи на бары 
    for container in ax1.containers:
        ax1.bar_label(container)
    
    plt.title('Медианное значения признака \n {}'.format(i), fontsize=14, pad=5)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.xlabel('Факт оттока')
    plt.ylabel('Медианное значение признака')

plt.tight_layout(w_pad=1)
plt.subplots_adjust(top=0.9) 
plt.show()
```

- средняя дополнительная выручка чуть меньше у тех, кто уходит - 115 против 158. Медиана немного сдвинула это значение в меньшую сторону - 103,8 против 149,8.  
- среднее количество месяцев до окончания срока действия абонемента меньше у уходяших и похоже на средний срок действия абномемента - 1,6 против 5,2 месяца. Медианное значение аналогично сроку действия абонемента выравнивает показатель - у уходящих 1 месяц, у оставшихся - 6. Это логично - люди стараются ходитьв в спорт-зал, если заплатили за более длительный абонемент

### Подгтовка выборок и стандартизация 

```python
# выделим из датасета столбцы с признаками 
X = df_gym.drop('Churn', axis=1)
# выделим вектор значений целевой переменной с фактом оттока 
y = df_gym['Churn']
```

```python
# разделим модель на обучающую и валидационную выборки 
# в соотношении 80 к 20 
# и зафиксируем ее состояние 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Так как целевая переменная представляет собой бинарную величину, то видим пики на 0 и 1. Относительная разница в верхушках на 0 и 1 для обучающей и валиадционной выборке примерно одинаковая, значит можем судить о высоком качестве разделения выборок 

```python
# дополнительный код с проверкой баланса классов и стратификацией 

# проверка долей 1 в тестовой и валидационной выбоках 
print('Доля 1 в обучающей выборке:',y_train.mean())
print('Доля 1 в тестовой выборке',y_test.mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    stratify=y, 
                                                    test_size=0.2, 
                                                    random_state=0)
print('Баланс классов после разделения выборок со стратификацией:')
print('Доля 1 в обучающей выборке:',y_train.mean())
print('Доля 1 в тестовой выборке',y_test.mean())
```

Поскольку прогнозирование оттока относится к бинарной классификации, то необходимо провести стандартизацию выборок 

```python
#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler() 
#обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(X_train)  
#display(X_train_st[:5])

#применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test) 
```

```python
model_LR = LogisticRegression(solver='liblinear', random_state=0)
model_LR.fit(X_train_st, y_train)
y_pred_LR = model_LR.predict(X_test_st)

#сделаем прогнозы
prob_LR = model_LR.predict_proba(X_test_st)[:,1]
```

### Обучение модели алгоритмом случайного леса 

```python
model_RFC = RandomForestClassifier(n_estimators = 100, random_state=0)
model_RFC.fit(X_train_st, y_train)
y_pred_RF = model_RFC.predict(X_test_st)

#сделаем прогнозы
prob_RFC = model_RFC.predict_proba(X_test_st)[:,1]
```

### Сравнение метрик 

```python
print('Метрики для модели логистической регрессии \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f} \n' \
      'F1:{:.2f} \n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y_test, y_pred_LR),
    precision_score (y_test, y_pred_LR),
    recall_score (y_test, y_pred_LR),
    f1_score(y_test, y_pred_LR),
    roc_auc_score(y_test, prob_LR)
))
print('')
print('Метрики для модели случайного леса \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f} \n' \
      'F1:{:.2f}\n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y_test, y_pred_RF),
    precision_score (y_test, y_pred_RF),
    recall_score (y_test, y_pred_RF),
    f1_score(y_test, y_pred_RF),
    roc_auc_score(y_test, prob_RFC)
))
```

Метрики показали, что чуть более точные результаты показывает модель логистической регрессии. Попробуем преобразовать исходную матрицу признаков и улучшить метрики обученных моделей

### Улучшение метрик обученных моделей после кластеризации

Попробуем улучшить метрики модели, после того, как провели кластеризацию. По сути мы введем в нашу матрицу признаков дополнительный признак - номер кластера. Будем использовать номера кластеров, полученные при разбиении на 5 кластеров.

```python
# выделим из датасета столбцы с признаками
X3 = df_gym.drop(['Churn',
              #    'gender', 
              #    'Phone',
              #   'Contract_period',
              #   'Avg_additional_charges_total',
                 'cluster_km4'], axis=1)
# выделим вектор значений целевой переменной с фактом оттока 
y3 = df_gym['Churn']

# разделим модель на обучающую и валидационную выборки 
# в соотношении 80 к 20 
# и зафиксируем ее состояние 
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, 
                                                        stratify=y,
                                                        test_size=0.2, 
                                                        random_state=0)

#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler() 
#обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X3_train_st = scaler.fit_transform(X3_train)  

# применяем стандартизацию к матрице признаков для тестовой выборки
X3_test_st = scaler.transform(X3_test) 

# Обучение модели алгоритмом логистической регрессии

model_LR3 = LogisticRegression(solver='liblinear', random_state=0)
model_LR3.fit(X3_train_st, y3_train)
y3_pred_LR = model_LR3.predict(X3_test_st)
prob3_LR = model_LR3.predict_proba(X3_test_st)[:,1]

# Обучение модели алгоритмом случайного леса 

model_RFC3 = RandomForestClassifier(n_estimators = 100, random_state=0)
model_RFC3.fit(X3_train_st, y3_train)
y3_pred_RF = model_RFC3.predict(X3_test_st)
prob3_RFC = model_RFC3.predict_proba(X3_test_st)[:,1]

print('Метрики для модели 3 логистической регрессии \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f} \n' \
      'F1:{:.2f}\n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y3_test, y3_pred_LR),
    precision_score (y3_test, y3_pred_LR),
    recall_score (y3_test, y3_pred_LR),
    f1_score(y3_test, y3_pred_LR),
    roc_auc_score(y3_test, prob3_LR)
))
print('')
print('Метрики для модели 3 случайного леса \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f}\n' \
      'F1:{:.2f}\n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y3_test, y3_pred_RF),
    precision_score (y3_test, y3_pred_RF),
    recall_score (y3_test, y3_pred_RF),
    f1_score(y3_test, y3_pred_RF),
    roc_auc_score(y3_test, prob3_RFC)
))
```

```python
df_gym['diff_freq'] =  df_gym['Avg_class_frequency_current_month'] - df_gym['Avg_class_frequency_total']
display(df_gym.head())
print('Корреляция между оттоком и модулем разницы частоты посещений всего и за месяц до оттока', 
     df_gym['Churn'].corr(df_gym['diff_freq']).round(decimals=2))
```

Видим высокую корреляцию между модулем разницы и фактом ооттока. 

```python
# выделим из датасета столбцы с признаками
X3 = df_gym.drop(['Churn',
                 'cluster_km5',
                 'cluster_km4'], axis=1)
# выделим вектор значений целевой переменной с фактом оттока 
y3 = df_gym['Churn']

# разделим модель на обучающую и валидационную выборки 
# в соотношении 80 к 20 
# и зафиксируем ее состояние 
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, 
                                                        stratify=y,
                                                        test_size=0.2, 
                                                        random_state=0)

#создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler() 
#обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X3_train_st = scaler.fit_transform(X3_train)  

# применяем стандартизацию к матрице признаков для тестовой выборки
X3_test_st = scaler.transform(X3_test) 

# Обучение модели алгоритмом логистической регрессии

model_LR3 = LogisticRegression(solver='liblinear', random_state=0)
model_LR3.fit(X3_train_st, y3_train)
y3_pred_LR = model_LR3.predict(X3_test_st)
prob3_LR = model_LR3.predict_proba(X3_test_st)[:,1]

# Обучение модели алгоритмом случайного леса 

model_RFC3 = RandomForestClassifier(n_estimators = 100, random_state=0)
model_RFC3.fit(X3_train_st, y3_train)
y3_pred_RF = model_RFC3.predict(X3_test_st)
prob3_RFC = model_RFC3.predict_proba(X3_test_st)[:,1]

print('Метрики для модели 3 логистической регрессии \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f} \n' \
      'F1:{:.2f}\n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y3_test, y3_pred_LR),
    precision_score (y3_test, y3_pred_LR),
    recall_score (y3_test, y3_pred_LR),
    f1_score(y3_test, y3_pred_LR),
    roc_auc_score(y3_test, prob3_LR)
))
print('')
print('Метрики для модели 3 случайного леса \n'\
      'Доля правильных ответов:{:.2f} \n' \
      'Точность ответов:{:.2f} \n' \
      'Полнота ответов:{:.2f}\n' \
      'F1:{:.2f}\n' \
      'ROC_AUC: {:.2f}'.format(
    accuracy_score(y3_test, y3_pred_RF),
    precision_score (y3_test, y3_pred_RF),
    recall_score (y3_test, y3_pred_RF),
    f1_score(y3_test, y3_pred_RF),
    roc_auc_score(y3_test, prob3_RFC)
))
```

Новый признак заметно улучшил обучение случайного леса. А логистическая реграссия осталась прежней.
